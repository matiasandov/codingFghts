```python
import collections
class Solution:
   
    def earliestAcq(self, logs: List[List[int]], n: int) -> int:
        """
        1. sort by timestamps
        2. crear uniones
            el truco es updatear el set de cada amigo en comun
            y parar si len(graph[persona]) == n
            
        """
        
        graph = collections.defaultdict(set)
        logs = sorted(logs, key = lambda i:i[0])

        #inicias el grafo con una conexion asi mismo, es la manera mas facil de iniciar el grafo
        for i in range(n):
            graph[i] = {i}

        
        for x,y,z in logs:
            graph[y] = graph[y].union(graph[z])
            
            #se va a actualizar el set de cada persona incluyendo z porque ya esta adentro
            for mutual in graph[y]:
                graph[mutual] = graph[y]

            if len(graph[y]) == n :
                return x
            

        return -1



```
